```python
import requests
import time
import matplotlib.pyplot as plt

url = "http://api.open-notify.org/iss-now.json"

def fetch_data():
    latitude_list, longitude_list, timestamp_list = [], [], []
    
    for _ in range(100):
        api_request = requests.get(url)
        data = api_request.json()
        
        latitude = float(data['iss_position']['latitude'])
        longitude = float(data['iss_position']['longitude'])
        timestamp = int(data['timestamp'])
        
        latitude_list.append(latitude)
        longitude_list.append(longitude)
        timestamp_list.append(timestamp)
        
        time.sleep(10)
        
    return latitude_list, longitude_list, timestamp_list

def graph_plotter(latitude_list, longitude_list):
    plt.figure(figsize=(10, 6))
    plt.plot(longitude_list, latitude_list, 'b.-')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')
    plt.title('Path of the International Space Station')
    plt.grid(True)
    plt.show()
    
latitude_list, longitude_list, _ = fetch_data()
graph_plotter(latitude_list, longitude_list)
```
